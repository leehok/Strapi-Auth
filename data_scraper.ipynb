```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_companies_of_the_Philippines#Notable_companies'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_ = "wikitable sortable")
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
ph_titles = table.find_all('th')
```

```python
ph_titles
```

```python
ph_table_titles = [title.text.strip() for title in ph_titles]

print(ph_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = ph_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]

  length = len(df)
  df.loc[length] =  individual_row_data
```

```python
df
```

```python
df.to_csv('Companies.csv', index = False)
```
